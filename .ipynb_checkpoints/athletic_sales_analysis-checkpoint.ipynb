```python
# Import Libraries and Dependencies
import pandas as pd
import datetime
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
df_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
df_2020.head(2)
```

```python
# Display the 2021 sales DataFrame
df_2021.head(2)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020.dtypes
```

```python
# Check the 2021 sales data types.
df_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df_2020_2021 = pd.concat([df_2020, df_2021], ignore_index=True)
```

```python
# Check if any values are null.
null_values = df_2020_2021.isnull()
```

```python
# Check the data type of each column
df_2020_2021.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
df_2020_2021['invoice_date'] = pd.to_datetime(df_2020_2021['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
df_2020_2021.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
prod_group = df_2020_2021.groupby(['region', 'state', 'city'])['product'].count().reset_index()
#Total_Products_Sold = 
prod_group.rename(columns={'product': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
prod_group.sort_values('Total_Products_Sold', ascending = False).head(5)                  
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
#location_pivot = df_2020_2021.pivot_table(index=['region', 'state', 'city'], values='product', aggfunc='count')
prod_pivot = pd.pivot_table(df_2020_2021, values='units_sold', index=['region', 'state', 'city'], aggfunc='count')
# Rename the "units_sold" column to "Total_Products_Sold"
prod_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
prod_pivot.sort_values('Total_Products_Sold', ascending = False).head(5) 
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
sales_group = df_2020_2021.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
#Total_Products_Sold = 
sales_group.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
sales_group.sort_values('Total Sales', ascending = False).head(5)   
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_pivot = pd.pivot_table(df_2020_2021, values='total_sales', index=['region', 'state', 'city'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
sales_pivot.sort_values('Total Sales', ascending = False).head(5)   
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailer_group = df_2020_2021.groupby(['retailer','region', 'state', 'city'])['total_sales'].sum().reset_index()
retailer_group.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the top 5 results.
retailer_group.sort_values('Total Sales', ascending = False).head(5)   
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_pivot = pd.pivot_table(df_2020_2021, values='total_sales', index=['retailer','region', 'state', 'city'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"

retailer_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the top 5 results.
retailer_pivot.sort_values('Total Sales', ascending = False).head(5)   
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
awf_sales = df_2020_2021[df_2020_2021['product'].str.contains("Women's Athletic Footwear")]
awf_sales.head(5)   
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
awf_group = awf_sales.groupby(['retailer','region', 'state', 'city'])['units_sold'].sum().reset_index()
awf_group.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the top 5 results.
awf_group.sort_values('Womens_Footwear_Units_Sold', ascending = False).head(5)   
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
awf_pivot = pd.pivot_table(awf_sales, values='units_sold', index=['retailer','region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
awf_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
awf_pivot.sort_values('Womens_Footwear_Units_Sold', ascending = False).head(5)   
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
awf_day_pivot = pd.pivot_table(awf_sales, values='total_sales', index=['invoice_date'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
awf_day_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Show the table.
awf_day_pivot
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

daily_awf_sales = awf_day_pivot.resample('D').sum()
awf_day_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_awf_sales.sort_values('Total Sales', ascending = False).head(10)   
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_awf_sales = awf_day_pivot.resample('W').sum()
# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_awf_sales.sort_values('Total Sales', ascending = False).head() 
```
